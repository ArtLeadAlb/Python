```python
# ### Importar e analisar tabelas

import pandas as pd
import numpy as np
import openpyxl
import matplotlib.pyplot as plt
import seaborn as sns


caminho = input('Copie o caminho da tabela desejada e aperte enter: ')
tabela = pd.read_excel(f"{caminho}")


def visualizacao_rapida(tabela):
    #rápida visualização dos dados

    print(tabela.head(5))
    print(tabela.tail(5))

def caracteristicas(tabela):
    #características da tabela
    print(tabela.shape)
    print(tabela.columns)
    print(tabela.dtypes)

def estatistica_basica(tabela):
    #resumo estatístico de cada coluna de números
    print(tabela.describe())

def verificar_duplicatas(tabela):
    #mostra dados duplicados
    print(tabela.duplicated())

def verificar_nulos(tabela):
    #mostra quantos registros são nulos em determinada coluna
    print(tabela.isnull().sum())
    #deleta linhas com valores nulos
    resolver = int(input('[1]-> Deletar linhas \n[2]-> Substituir pela média aritimética \n[3]-> Substituir pela mediana \nO que deseja fazer com os valores nulos? \n'))
    
    if resolver == 1:
        tabela.dropna()
                         
    elif resolver == 2:
    # Preencher valores nulos da coluna 'Idade' com a média da coluna
        coluna = input('Nome da Coluna: ')
        media_idade = tabela[coluna].mean()
        tabela = tabela[coluna].fillna(media_idade, inplace = True)

    elif resolver == 3:
        # Preencher valores nulos da coluna 'Salario' com a mediana da coluna
        coluna = input('Nome da Coluna: ')
        mediana_salario = tabela[coluna].median()
        tabela = tabela[coluna].fillna(mediana_salario, inplace = True)

def  info_corr(tabela):
    print(tabela.info())
    #correlação entre as colunas
    print(tabela.corr())

def grafico(tabela):
    import matplotlib.pyplot as plt
    import seaborn as sns
    #plotar o gráfico
    #plt.scatter(x, y, bins = 1, alpha = 0.8) #bins->agrupamento de dados  alpha->opacidade
    eixo_x = input('Eixo x: ')
    eixo_y = input('Eixo y: ')
    plt.scatter(x=tabela[eixo_x], y=tabela[eixo_y])
    plt.xlabel(f'{eixo_x}')
    plt.ylabel(f'{eixo_y}')
    plt.title(f'{eixo_x} x {eixo_y}')
    #plt.legend()
    #mostrar o gráfico
    plt.show()
    

def sm_RegressaoLinear(tabela):
    import statsmodels.api as sm

    #Regressão linear       obs:  funcioina sem passar x como np.array e depois usar x=x.reshape[-1,1] ???
    coluna1 = input("nome da coluna 1: ")
    coluna2 = input("nome da coluna 2: ")
    x = tabela[coluna1]
    y = tabela[coluna2]

    #adicionar constante para dar certo
    x = sm.add_constant(x)

    #criar modelo
    modelo = sm.OLS(y, x)
    resultado = modelo.fit()

    #Mostra o resultado do modelo
    print('\n======================================= STATSMODELS =======================================\n')
    print(resultado.summary()) #R-squared determina a correlação entre as variáveis. 0->1
    print('\n======================================= TUDO CERTO! =======================================\n')



```

```python
while True:
    print('\nMenu de opções:\n[1]-> Breve visualização da tabela \n[2]-> Características básicas da tabela \n[3]-> Estatísticas básicas da tabela \n[4]-> Verificar duplicatas \n[5]-> Verificar e corrige nulos \n[6]-> Informações e Correlações \n[7]-> Gerar gráfico \n[8]-> Regressão Linear \n[0]-> Sair')
    opcao = int(input('Selecione uma opção: '))
    if opcao == 1:
        print('\n')
        visualizacao_rapida(tabela)
        print('\n')
    elif opcao == 2:
        print('\n')
        caracteristicas(tabela)
        print('\n')
    elif opcao == 3:
        print('\n')
        verificar_duplicatas(tabela)
        print('\n')
    elif opcao == 4:
        print('\n')
        verificar_nulos(tabela)
        print('\n')
    elif opcao == 5:
        print('\n')
        verificar_nulos(tabela)
        print('\n')
    elif opcao == 6:
        print('\n')
        info_corr(tabela)
        print('\n')
    elif opcao == 7:
        print('\n')
        grafico(tabela)
        print('\n')
    elif opcao == 8:
        print('\n')
        sm_RegressaoLinear(tabela)
        print('\n')
    elif opcao == 0:
        break
```
